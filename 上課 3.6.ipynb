```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,10,100)
```

```python
y = 2*x +3 + np.random.randn(100)
```

```python
plt.scatter(x,y)
```

```python
def f(x,a,b):
    return a*x + b
```

```python
from scipy.optimize import curve_fit 
```

```python
popt, pcov = curve_fit(f,x,y)  # popt是參數，就是上面的a,b   
```

```python
a,b = popt
```

```python
plt.figure(figsize=(15,9))
plt.scatter(x,y)
plt.plot(x,a*x+b,"r")
plt.title("hi",size=60)
plt.xlabel("x",size = 20)
plt.ylabel("y",size = 20)

plt.axis([0, 6, 0, 20]) # 可用來放大縮小座標軸
```
