```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.array([[-3,2],[-6,5],[3,-4],[2,-8]])    # 把list轉成array比較好
y = np.array([1,1,2,2])
```

```python
x[:,0] # 先index，再columns。和 pandas很像
```

# SVM （ 監督式

```python
from sklearn.svm import SVC
```

```python
clf = SVC()   #開一個分類器
```

```python
clf.fit(x,y)
```

```python
clf.predict([-1,3])
```

 # 繼續

```python
gd = np.array([(i,j) for i in np.arange(-10,4,0.6) for j in np.arange(-4,8,0.6)])
```

 # K-Mean （非監督

```python
x = np.random.rand(100,2)
```

```python
plt.scatter(x[:,0],x[:,1],s=30)
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(3)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.figure(figsize=(20,10))


plt.scatter(x[:,0],x[:,1],s=30,c = clf.labels_)
plt.grid()
```

```python
gd = np.array([(i,j) for i in np.arange(-0.2,1.2,0.05)
              for j in np.arange(-0.2,1.2,0.05)])
```

```python
gcd = clf.predict(gd)
```

```python
plt.scatter(gd[:,0],gd[:,1],c=gcd)
```

```python
# fit 是訓練 ， predict就是預測！
```
